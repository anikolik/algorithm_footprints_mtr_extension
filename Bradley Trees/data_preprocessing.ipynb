### Preference data

```python
def pairwise_comparison(x, y):
  """ Function to compare two values. """
  if x < y:
      return 1
  elif x == y:
      return 0
  else:
      return -1

def prepare_preference_data(file_path):
  # Read algorithm performance data
  y = pd.read_csv(file_path, index_col=["f_id", "i_id"])
  print("Preview performance data:")
  print(y.head())
  # Compare algorithm performance
  preference_df = pd.DataFrame(index=y.index)
  for combination in itertools.combinations(y.columns, 2):
    preference_df[f'{combination[0]}_{combination[1]}'] = y.apply(lambda row: pairwise_comparison(row[combination[0]], row[combination[1]]), axis=1)

  return preference_df

preference_df = prepare_preference_data(file_path=f"{DATA_PATH}/y_train.csv")
print("\nPreview preference data:")
print(preference_df.head())
print(preference_df.shape)
# save
preference_df.to_csv(f"{SAVE_PATH}/y_train_preference.csv", index=False)
```
